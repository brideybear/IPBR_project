```python
codon_chart = {
    'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',
    'UAU': 'Y', 'UAC': 'Y', 'UAA': 'STOP', 'UAG': 'STOP',
    'UGU': 'C', 'UGC': 'C', 'UGA': 'STOP', 'UGG': 'W',
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I', 'AUG': 'M',  # AUG is also START
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
    'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
}
#codon chart library for translation later
```

```python
from Bio import Blast
from Bio import SeqIO
from Bio.Blast import NCBIWWW
from Bio.Blast import NCBIXML
```

```python
path="/Users/Bridey/Downloads/Mystery_Gene.txt"
#Defining the local file path for the gene sequence
```

```python
with open(path) as geneSeq:
    sequence = geneSeq.read()
RNASeq=sequence.replace('T','U')
#Transcription: Loading the sequence, then converting from DNA to RNA
```

```python
#Define the function to translate RNA sequence into an amino acid string
def translate_rna(RNASeq, codon_chart):
    codons = []
    amino_acids = []
    
    #Goes through the sequence in sections of 3 to extract/separate the codons
    for i in range(0, len(RNASeq) - (len(RNASeq) % 3), 3):
        codon = RNASeq[i:i+3]
        codons.append(codon)

        #Looks up codon in codon chart and handles unknown codons by marking an X or STOP codons by stopping the loop.
        if codon in codon_chart:
            amino_acid = codon_chart[codon]
            if amino_acid != "STOP":
                amino_acids.append(amino_acid)
        else:
            amino_acids.append('X')
    return "".join(amino_acids)

    print(f"Codons:      {' '.join(codons)}")
    print(f"Translation: {' '.join(amino_acids)}")
```

```python
#Execute the translation and show the resulting amino acid sequence
proteinSeq = translate_rna(RNASeq, codon_chart)
print(proteinSeq)
```

```python
#Execute a protein BLAST search against the nr (non-redundant) database
#This sends the sequence to NCBI servers and may take a while
result_stream = NCBIWWW.qblast("blastp", "nr", proteinSeq)
```

```python
#Use NCBIXML to read through the XML results returned from the BLAST search
blast_result = NCBIXML.read(result_stream)
```

```python
#Retrieve and display the title of the closest match
blast_result.alignments[0].title
```
